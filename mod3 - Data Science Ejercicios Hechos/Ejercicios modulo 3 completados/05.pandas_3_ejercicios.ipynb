### Ejercicio 18:
- Obtener el tipo de elementos de cada columna y convierte los que sean "int64" a "int32". Evalua la memoria que consume antes y después de convertir las columnas y comenta la diferencia.

```python
# 1. Obtener el tipo de elementos de cada columna
tipos_antes = df.dtypes
print("Tipos de elementos antes de la conversión:")
print(tipos_antes)

# 2. Evaluar la memoria consumida antes de la conversión
memoria_antes = df.memory_usage(deep=True).sum()
print(f"\nMemoria consumida antes de la conversión: {memoria_antes} bytes")

# 3. Convertir las columnas de tipo "int64" a "int32"
df = df.astype({col: 'int32' for col in df.select_dtypes('int64').columns})

# 4. Evaluar la memoria consumida después de la conversión
memoria_despues = df.memory_usage(deep=True).sum()
print(f"\nMemoria consumida después de la conversión: {memoria_despues} bytes")

# 5. Comentar la diferencia
diferencia = memoria_antes - memoria_despues
print(f"\nDiferencia en memoria consumida: {diferencia} bytes")
```

### Ejercicio 20:
- Agrega una nueva columna al DataFrame que se llame: "Rango edad" y va a categorizar de la siguiente manera:
    - Niño < 11 años
    - Joven < 18 años
    - Adulto < 50 años
    - Mayor >= 50 años

```python
def categorizar_edad(age):
    if age < 11:
        return 'Niño'
    elif age < 18:
        return 'Joven'
    elif age < 50:
        return 'Adulto'
    else:
        return 'Mayor'

# Aplicar la función y crear la nueva columna "Rango edad"
df['Rango edad'] = df['Age'].apply(categorizar_edad)
df.head()
```

### Ejercicio 26:
- Divide la columna name en las siguientes columnas:
    - Mrs/Miss
    - Apellido
    - Nombre(s)
    
**Nota: Ignora el resto de información.**

```python
def dividir_nombre(name):
    partes = name.split(', ')
    apellido = partes[0]
    resto = partes[1].split('. ')
    mrs_miss = resto[0]
    nombres = resto[1]
    return pd.Series([mrs_miss, apellido, nombres])

# Aplicar la función y crear las nuevas columnas
df[['Mrs/Miss', 'Apellido', 'Nombre(s)']] = df['Name'].apply(dividir_nombre)

df.head()

```
